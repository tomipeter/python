## 엑셀 파일 Merge => Pivot 돌리기

```python
import pandas as pd
filename = r"C:\Users\UBASE\Downloads\UBASE HC.xlsx"
xls_file2 = pd.ExcelFile(filename)

df = xls_file2.parse('2016',skiprows=2,ignore_index=True)
df2 = xls_file2.parse('2017',skiprows=2,ignore_index=True)
df.rename(columns=
          {'English':'NT_English',
           'Japanese':'NT_Japanese',
           'Korean':'NT_Korean',
           'Mandarin':'NT_Madarin',
           'Other':'NT_Other',
           'English.1':'T_English',
           'Japanese.1':'T_Japanese',
           'Korean.1':'T_Korean',
           'Mandarin.1':'T_Mandarin',
           'Other.1':'T_Other',
                    
          },inplace=True)
df2.rename(columns=
          {'English':'NT_English',
           'Japanese':'NT_Japanese',
           'Korean':'NT_Korean',
           'Mandarin':'NT_Madarin',
           'Other':'NT_Other',
           'English.1':'T_English',
           'Japanese.1':'T_Japanese',
           'Korean.1':'T_Korean',
           'Mandarin.1':'T_Mandarin',
           'Other.1':'T_Other',
                    
          },inplace=True)

frames = [df,df2]
result = pd.concat(frames)
result
result_excel = pd.melt(result.reset_index(), id_vars=['Week'],value_vars=['NT_Korean','NT_Japanese','NT_Mandarin',
                                                       'T_Korean','T_Japanese','T_Mandarin'])

result_excel.to_csv("C:/UBASE/Airbnb/Data/HeadCount/Airbnb_HeadCount.csv")
```
